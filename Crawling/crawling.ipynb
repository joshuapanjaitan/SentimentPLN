# Crawling
Crawling adalah proses di belakang sebuah search engine, yang bertugas menelusuri World Wide Web secara terstruktur dan dengan etika-etika tertentu.
Disini web yang di-crawl adalah twitter dengan mencari tweet yang sesuai dengan topik yang sudah didefinisikan diatas.

Aturan Crawl : Hanya tweet berbasa indonesia, Mengabaikan tweet yang ditulis oleh bot

```python
pages = []
for x in range(len(topik)):
    startDate = datetime.datetime(2021, 1, 20, 0, 0, 0)  # set tanggal Mulai
    endDate = datetime.datetime(2021, 1, 25, 0, 0, 0)  # Set Tanggal Selesai
    for tweet in tweepy.Cursor(api.search, q=topik[x]).items(10): #crawl, items(1), crawl max 1 tweet/topik
        box = []
        status = api.get_status(tweet.id, tweet_mode="extended")
        text = status.full_text
        lang = tweet.lang
        if lang == 'in':  # cek bahasa, hanya indo yg masuk
            if len(text) != 0:
                cek = text[0] + text[1]
                # if cek != 'RT':  # cek retweet
                if tweet.created_at < endDate and tweet.created_at > startDate:  # ambil tweet dari rentang waktu tertentu
                    dateTimeObj = tweet.created_at
                    timestampStr = dateTimeObj.strftime("%d-%b-%Y")
                    box.append(timestampStr)
                    box.append(text)
        if len(box) != 0:
            pages.append(box)
            box = []
```

```python
with open("new-crawl.csv", "w") as f:
    writer = csv.writer(f)
    for i in range(len(pages)):
        v = str(pages[i][1])
        g = str(v.encode('unicode-escape'))
        pages[i][1] = g
        writer.writerow(pages[i])
```
