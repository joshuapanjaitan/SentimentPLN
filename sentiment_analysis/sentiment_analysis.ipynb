# Klasifikasi Sentiment Kedalam Kelas Positive dan Negative 
Proses ini merupakan proses utama klasifikasi sentiment kedalam kelas positive dan negative dengan menggunakan algoritma Naive Bayes

```python
import nltk
import csv
import pandas as pd
```

```python
#baca file tweet.csv, merupakan tweet yang sudah dibersihkan dari proses cleaning
data = pd.read_csv('tweet.csv', usecols=[0, 1], names=['date', 'tweet'])
sentimentSign = []
```

```python
# Fungsi untuk membaca sebuah file
def openText(name):
    file = open(name, encoding="utf-8")
    sentences = []
    for line in file.read().splitlines():
        words = line.split()
        sentences.append(words)
    return sentences
```

```python
#Fungsi untuk mendapatkan fitur
def get_features(wordlist):
    wordlist = nltk.FreqDist(wordlist)
    word_features = wordlist.keys()
    return word_features
```

```python
#fungsi untuk membaca semua tweet
def get_tweets(tweets):
    all_words = []
    for (words, sentiment) in tweets:
        all_words.extend(words)
    return all_words
```

```python
#fungsi untuk mengembalikan fitur
def take_features(document):
    features = {}
    for word in word_features:
        features['contains(%s)' % word] = (word in document)
    return features
```

```python
# Load Training set into array
file_negatif = openText('dataset/sentences_negatif.txt')
file_positif = openText('dataset/sentences_positif.txt')
```

```python
tweets = [] #simpan tweet
for words in file_positif:  # label untuk positif 1
    tweets.append((words, 1))


for words in file_negatif:  # label untuk negative 0
    tweets.append((words, 0))
```

```python
word_features = get_features(get_tweets(tweets)) #feature extraction
```

```python
# build training sets using word_features
training_set = nltk.classify.apply_features(take_features, tweets)
```

```python
# build classifier model
classifier = nltk.NaiveBayesClassifier.train(training_set)
print("Model Rendered")
classifier.show_most_informative_features()
```

```python
#proses klasifikasi
for line in data['tweet']:
    tweet = str(line)
    ress = classifier.classify(take_features(tweet.split())) #klasifiasi tweet
    if ress == 1:
        sentimentSign.append('Positive')
    elif ress == 0:
        sentimentSign.append('Negative')
```

```python
data['polarity'] = sentimentSign
data.to_csv('hasil.csv')
pd.set_option('display.max_colwidth', -1)
```
